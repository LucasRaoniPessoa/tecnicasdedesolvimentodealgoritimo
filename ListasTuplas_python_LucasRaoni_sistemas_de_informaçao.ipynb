```python
#ATIVIDADE2

# Escreva um programa que aplica a função módulo a todos os elementos e uma lista.

lista_com_negativos = [10, -20, 30, -40, -5]

lista_com_modulo = [abs(x) for x in lista_com_negativos]

print("Lista original:", lista_com_negativos)
print("Lista com módulo:", lista_com_modulo)
```

```python
#ATIVIDADE3

# Escreva um programa que apaga todos os elementos negativos de uma lista.

lista_com_negativos = [10, -20, 30, -40, -5]

lista_sem_negativos = [x for x in lista_com_negativos if x >= 0]

print("Lista original:", lista_com_negativos)
print("Lista sem negativos:", lista_sem_negativos)
```

```python
#ATIVIDADE4

# Faça um programa que solicita uma quantidade indefinida de notas de uma prova (deve parar quando um número negativo for digitado).
# Armazena em uma lista.
# Após isso, o programa deve mostrar a média das notas.

notas = []
while True:
    nota = float(input("Digite uma nota (ou um número negativo para parar): "))
    if nota < 0:
        break
    notas.append(nota)

print("Notas inseridas:", notas)
```

```python
#ATIVIDADE5

# Faça um programa que solicita uma quantidade indefinida de números positivos
# (deve parar quando um número negativo for digitado).
# Armazene-os em duas listas:
# uma para os números pares;
# outra para os números ímpares.
# A seguir, mostre a porcentagem de pares e ímpares digitados.

numeros_pares = []
numeros_impares = []
total_numeros = 0

while True:
    numero = float(input("Digite um número positivo (ou um número negativo para parar): "))
    if numero < 0:
        break
    total_numeros += 1
    if numero % 2 == 0:
        numeros_pares.append(numero)
    else:
        numeros_impares.append(numero)

print("Números pares:", numeros_pares)
print("Números ímpares:", numeros_impares)
print("Total de números digitados:", total_numeros)
```

```python
#atividade7

pessoas = []
for i in range(5):
    print(f"Digite os dados da pessoa {i+1}:")
    altura = float(input("Altura: "))
    peso = float(input("Peso: "))
    pessoas.append({"altura": altura, "peso": peso})

tem_duplicado = False
for i in range(len(pessoas)):
    for j in range(i + 1, len(pessoas)):
        if pessoas[i]["altura"] == pessoas[j]["altura"] and pessoas[i]["peso"] == pessoas[j]["peso"]:
            tem_duplicado = True
            break
    if tem_duplicado:
        break

if tem_duplicado:
    print("\nPelo menos duas pessoas têm a mesma altura e mesmo peso.")
else:
    print("\nNenhuma pessoa tem a mesma altura e mesmo peso que outra.")
```

```python
#Atividade8

pessoas = []
while True:
    nome = input("Digite o nome da pessoa (ou deixe vazio para parar): ")
    if nome == "":
        break
    idade = int(input(f"Digite a idade de {nome}: "))
    pessoas.append({"nome": nome, "idade": idade})

if pessoas:
    pessoa_mais_velha = max(pessoas, key=lambda x: x['idade'])
    print(f"\nA pessoa mais velha é: {pessoa_mais_velha['nome']}")
else:
    print("\nNenhum dado de pessoa foi inserido.")
```

```python
alunos_db = [
    {"matricula": "2023001", "nome": "Lucas raoni", "curso": "sistemas de informaçao"},
    {"matricula": "2023002", "nome": "Ana Castela", "curso": "Engenharia de software"},
    {"matricula": "2023003", "nome": "Gustavo Pereira", "curso": "Analise e desenvolvimento de sistemas"},
]
print("1. Sistema simplificado de cadastro de aluno")
print("-" * 40)

#1. Adicionar novo aluno (create)
print("-> Adicionado novo aluno...")
novo_aluno = {"matricula": "2023004", "nome": "Joao Gustavo", "curso": "Sistema de informaçao"}
alunos_db.append(novo_aluno) #adiciona o dicionario (novo_aluno) ao
print(f"Aluno {novo_aluno['nome']} adicionado com sucesso.")

#2 Buscar aluno por matricula (READ)
matricula_busca = "2023002"
aluno_encontrado = next((aluno for aluno in alunos_db if aluno["matricula"] == matricula_busca), None)

if aluno_encontrado:
  print(f"\nAluno com matricula (Matricula: {matricula_busca}):")
  print(f"Nome: {aluno_encontrado['nome']}, curso: {aluno_encontrado['curso']}")
else:
  print(f"\nAluno com matricula {matricula_busca} nao encontrado")

# 3. Atualizar curso do aluno (UPDATE)
matricula_update = "2023003"
for aluno in alunos_db:
    if aluno["matricula"] == matricula_update:
      aluno["curso"] = "segurança da informaçao"
      print(f"\nCurso do aluno {aluno['nome']} atualizado para: {aluno['curso']}")
      break
print("-" * 40)

# 4. Excluir aluno (DELETE)
matricula_delete = "2023001"
alunos_db = [aluno for aluno in alunos_db if aluno["matricula"] != matricula_delete]
print(f"\nAluno com matricula {matricula_delete} excluido com sucesso.")

print("\nAlunos restantes:")
for aluno in alunos_db:
    print(f"Matricula: {aluno['matricula']}, Nome: {aluno['nome']}, Curso: {aluno['curso']}")
```
